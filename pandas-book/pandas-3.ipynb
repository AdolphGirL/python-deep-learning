```python
# https://data.taipei/api/getDatasetInfo/downloadResource?id=2f238b4f-1b27-4085-93e9-d684ef0e2735&rid=f47c1289-9e2f-49aa-a328-37708402a9ac
base_url = "https://data.taipei/api/getDatasetInfo/downloadResource?id={}&rid={}"
_id = "2f238b4f-1b27-4085-93e9-d684ef0e2735"
_rid = "f47c1289-9e2f-49aa-a328-37708402a9ac"
csv_url = base_url.format(_id, _rid)
csv_url
```

```python
df_raw = pd.read_csv(csv_url, encoding='big5')
df_raw
```

```python
df = df_raw.copy()
df
```

```python
# 去除區序代碼
df['區序'] = df['區序'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
df.head()
```

```python
df = (df[df['性別'].isin([1, 2])]
      .groupby(['區序', '性別'])[['死亡人數', '受傷人數']]
      .sum()
      .reset_index()
      .sort_values('受傷人數'))
df
```

```python
df['性別'] = df['性別'].apply(lambda x: '男性' if x == 1 else '女性')
df
```
