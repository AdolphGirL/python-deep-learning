```python
titanic_info.dropna(axis=1)
```

```python
def age_label(row):
#     print(row)
    age = row['Age']
    if pd.isnull(age):
        return 'unknow'
    elif age < 18:
        return 'minor'
    else:
        return 'adult'
    
age_lables = titanic_info.apply(age_label, axis=1)
age_lables
```

```python
titanic_info['age_labels'] = age_lables
age_group_su = titanic_info.pivot_table(index='age_labels', values='Survived', aggfunc=np.mean)
age_group_su
```
